```python
import pandas as pd
```

```python
matches = pd.read_csv("matches.csv")
matches.head(10)
```

```python
matches.info()
```

```python
matches.columns
```

```python
matches.columns = ['Scorecard', 'Team_1', 'Team_2', 'Winner', 'Margin', 'Ground',
       'Match_Date']
matches.head()
```

Convert to Date

```python
matches.Team_1
```

```python
matches.Match_Date = pd.to_datetime(matches.Match_Date,format='%b %d, %Y',errors = 'coerce')
```

Pre - Processing

```python
matches.info()
```

```python
matches.head()
```

```python
matches['Match_Year'] = matches.Match_Date.dt.year
```

```python
matches.head()
```

```python
matches.Match_Year.describe()
```

```python
matches_07 = matches[matches.Match_Year >= 2007]
matches.head()
```

```python
matches_07.count()
```

```python
matches_07.Team_1.unique()
```

```python
major_countries = ['Afghanistan', 'Australia', 'Bangladesh', 'England', 'India', 'New Zealand', 'Pakistan', 'South Africa	', 'Sri Lanka', 'West Indies', 'Zimbabwe']

matches_07 = matches_07[matches_07.Team_1.isin(major_countries)]
matches_07 = matches_07[matches_07.Team_1.isin(major_countries)]
matches_07.count()
```

```python
matches_07.head()
```

```python
matches_07['team_1_win'] = matches_07.Team_1 == matches_07.Winner
matches_07['team_2_win'] = matches_07.Team_2 == matches_07.Winner
matches_07.head(15)
```

Aggregating the Data

```python
matches_07['team_1_win'] = matches_07.Team_1 == matches_07.Winner
matches_07['team_2_win'] = matches_07.Team_2 == matches_07.Winner
matches_07.head()
```

```python
matches_07.groupby('Team_1').agg({'Scorecard': 'count', 'team_1_win': 'sum'})
```

```python
team_1 = matches_07.groupby(['Team_1','Match_Year']).agg({'Scorecard': 'count', 'team_1_win': 'sum'}).reset_index()
team_2 = matches_07.groupby(['Team_2','Match_Year']).agg({'Scorecard': 'count', 'team_2_win': 'sum'}).reset_index()
team_1.head()
```

```python
team_2.head()
```

```python
team_1 = team_1.rename(columns = {'Team_1': 'team', 'Scorecard': 'total_matches'})
team_2 = team_2.rename(columns = {'Team_2': 'team', 'Scorecard': 'total_matches'})
```

```python
team_1.head()
```

```python
team_2.head()
```

```python
team_info = pd.merge(team_1, team_2, on = ['team', 'Match_Year'], how='outer')
print(team_info.count())
team_info.head()
```

```python
team_info = team_info.fillna(0)
team_info.head()
```

```python
team_info['total_matches'] = team_info['total_matches_x'] + team_info['total_matches_y']
team_info['total_wins'] = team_info['team_1_win'] + team_info['team_2_win']

team_info.tail(30)
```

```python
team_info['win_perc'] = 100* team_info['total_wins']/team_info['total_matches']

team_info.head()
```

```python
team_info_x = team_info[team_info.team.isin(['India','Pakistan', 'Australia', 'England'])]
```

```python
team_info_x.pivot(index='Match_Year', columns='team', values='win_perc').plot.line()
```
